```python
STN1590 = pd.read_csv("produkt_klima_tag_19631001_20181231_01590.txt", sep = ';', engine = 'python')
STN2629 = pd.read_csv("produkt_klima_tag_19471111_20181231_02629.txt", sep = ';', engine = 'python')
STN13670 = pd.read_csv("produkt_klima_tag_20070531_20181231_13670.txt", sep = ';', engine = 'python')
STN5064 = pd.read_csv("produkt_klima_tag_20041201_20181231_05064.txt", sep = ';', engine = 'python')
STN1078 = pd.read_csv("produkt_klima_tag_19520101_20181231_01078.txt", sep = ';', engine = 'python')
STN1303 = pd.read_csv("produkt_klima_tag_19350101_20181231_01303.txt", sep = ';', engine = 'python')
```

```python
STN = STN1590.append(STN2629).append(STN13670).append(STN5064).append(STN1078).append(STN1303)
```

```python
STN['MESS_DATUM'] = pd.to_datetime(STN['MESS_DATUM'], format = '%Y%m%d')
```

```python
STN_Final = STN[['MESS_DATUM',' RSK',' TMK']]
```

```python
STN_Final.set_index('MESS_DATUM',inplace=True)
```

```python
STN_1518 = STN_Final.loc['2016':'2018']
```

```python
STN_1518_byMonth = STN_1518[[' RSK',' TMK']].resample('M').mean()
```

```python
sns.set(rc={'figure.figsize':(18,4)})

STN_1518_byMonth[[' RSK',' TMK']].plot(linewidth=0.5)
```
